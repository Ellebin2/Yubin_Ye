```python
import time
import pandas as pd
import numpy as np

CITY_DATA = { 'chicago': 'chicago.csv',
              'new york': 'new_york_city.csv',
              'washington': 'washington.csv' }

def get_filters():
    print('Hello! Let\'s explore some US bikeshare data!')
    # TO DO: get user input for city (chicago, new york city, washington). 
    #HINT: Use a while loop to handle invalid inputs
    
    city = input('\nWhich city you`d like to enter ?.\n')
    while city not in CITY_DATA.keys():
        city = input('\n only include data of chicago, new york city or washington.\n')

    # TO DO: get user input for month (all, january, february, ... , june)
    month = input('\nWhich month you`d like to enter?.\n')
    month_list = ['1','2','3','4','5','6']
    while month not in month_list:
        month = input('\n beyound month range,try1,2,3,4,5,6.\n')

    # TO DO: get user input for day of week (all, monday, tuesday, ... sunday)
    day = input('\nWhich day you like to enter?.\n')
    day_list = str(list(np.arange(1,32)))
    while day not in day_list:
        day = input('\n beyound date range,try:1~31.\n')
    
    print('-'*40)
    return city, month, day

def get_city(city):
    city_low = city.lower()
    if city_low == 'new york':
        df = pd.read_csv('new_york_city.csv', error_bad_lines=False)
    elif city_low == 'washington':
        df = pd.read_csv('washington.csv', error_bad_lines=False)
        print(df.columns)
    elif city_low == 'chicago':
        df = pd.read_csv('chicago.csv', error_bad_lines=False)
    else:
        'no such city' 
    return(df)

def get_time(df,month,day):
    m = int(month)
    d = int(day)
    df['start time'] = pd.to_datetime(df['Start Time'])
    df['month'] = df['start time'].dt.month
    df['day_of_week'] = df['start time'].dt.weekday
    df['day'] = df['start time'].dt.day
    df['hour'] = df['start time'].dt.hour
    
    hour_zip = zip(df['month'],df['day'],df['hour'])

    hours = []
    for i in hour_zip:
        if (i[0] == m) & (i[1] == d):
                hours.append(i[2])

    hour_count = {}
    for i in hours:
        if hours.count(i)>1:
            hour_count[i] = hours.count(i)

    a = max(zip(hour_count.values(),hour_count.keys()))
    b = '(count,hour): {}'.format(a)
    print(b)
    
def time_stats(df):          
    """Displays statistics on the most frequent times of travel."""

    print('\nCalculating The Most Frequent Times of Travel...\n')
    start_time = time.time()

    # TO DO: display the most common month
    df['start time'] = pd.to_datetime(df['Start Time'])
    df['month'] = df['start time'].dt.month
    pop_month = df['month'].mode()[0]
    p_m = 'popular_month: {}'.format(pop_month)
    print(p_m)

    # TO DO: display the most common day of week
    df['day_of_week'] = df['start time'].dt.weekday
    pop_day = df['day_of_week'].mode()[0]
    p_w = 'popular_weekday: {}'.format(pop_day)
    print(p_w)

    # TO DO: display the most common start hour
    df['hour'] = df['start time'].dt.hour
    pop_hour = df['hour'].mode()[0]
    p_h = 'popular_hour: {}'.format(pop_hour)
    print(p_h)

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
#     return pop_month,pop_day,pop_hour

def station_stats(df):
    """Displays statistics on the most popular stations and trip."""

    print('\nCalculating The Most Popular Stations and Trip...\n')
    start_time = time.time()

    # TO DO: display most commonly used start station
    pop_start = df['Start Station'].mode()[0]
    p_s = 'popular_start_station: {}'.format(pop_start)
    print(p_s)

    # TO DO: display most commonly used end station
    pop_end = df['End Station'].mode()[0]
    p_e = 'popular_end_station: {}'.format(pop_end)
    print(p_e)

    # TO DO: display most frequent combination of start station and end station trip
    pop_start_count = df['Start Station'].value_counts()
    start_index = list(pop_start_count.index)
    start_count = list(pop_start_count)

    pop_end_count = df['End Station'].value_counts()
    end_index = list(pop_end_count.index)
    end_count = list(pop_end_count)

    index = zip(start_index,end_index)
    count = zip(start_count,end_count)

    group = []
    for i,j in index:
        station = 'start: '+i,'end: '+j
        group.append(station)

    counts = []
    for i,j in count:
        sum_ = i+j
        counts.append(sum_)

    d = dict(zip(group,counts))
    sorted(d.items())

    max_group = max(zip(d.values(),d.keys()))
    m_g = 'popular_station_group: {}'.format(max_group)
    print(m_g)

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)


def trip_duration_stats(df):
    """Displays statistics on the total and average trip duration."""

    print('\nCalculating Trip Duration...\n')
    start_time = time.time()

    # TO DO: display total travel time
    trip_sum = df['Trip Duration'].sum()
    t_s = 'sum of trip: {}'.format(trip_sum)
    print(t_s)

    # TO DO: display mean travel time
    trip_mean = df['Trip Duration'].mean()
    t_m = 'mean of trip: {}'.format(trip_mean)
    print(t_m)

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)


def user_stats(df):
    """Displays statistics on bikeshare users."""

    print('\nCalculating User Stats...\n')
    start_time = time.time()

    # TO DO: Display counts of user types
    user_types = df['User Type'].value_counts()
    u_t = '{}'.format(user_types)
    print(u_t)  

    # TO DO: Display counts of gender
    try:
        Genders = df['Gender'].value_counts()
        g_d = '{}'.format(Genders)
        print(g_d)
    except:
        KeyError: 'Gender'

    # TO DO: Display earliest, most recent, and most common year of birth
    try:
        early_birth = df['Birth Year'].min()
        late_birth = df['Birth Year'].max()
        pop_birth = df['Birth Year'].mode()[0]

        e_b = 'earliest_birth: {}'.format(early_birth)
        l_b = 'latest_birth: {}'.format(late_birth)
        p_b = 'most_popular_birth: {}'.format(pop_birth)

        print(e_b)
        print(l_b)
        print(p_b)
    except:
        KeyError: 'Birth Year'

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)

def main():
    while True:
        city, month, day = get_filters()
        df = get_city(city)
        get_time(df,month,day)

        time_stats(df)
        station_stats(df)
        trip_duration_stats(df)
        user_stats(df)

        restart = input('\nWould you like to restart? Enter yes or no.\n')
        if restart.lower() != 'yes':
            break

if __name__ == "__main__":
    main()
```
